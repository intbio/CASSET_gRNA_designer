## Targets for split-system

```python
%%time
pathogen_dict = {'sars-cov-2':'NC_045512.2'}
pathogen_genomes = get_pathogen_genomes(pathogen_dict)

n_gene = pathogen_genomes['sars-cov-2'][28274:29533]
n_gene_coordinates = range(28274,29533)

e_gene = pathogen_genomes['sars-cov-2'][26245:26472]
e_gene_coordinates = range(26245,26472)

pathogen_dict.update({'n_gene':'n_gene', 'e_gene':'e_gene'})
pathogen_genomes.update({'n_gene':n_gene, 'e_gene':e_gene})

pathogen_pam = get_pathogen_pam(pathogen_dict, pathogen_genomes, cas_info)
```

```python
%%time

pathogen_pam_pairs = {}
pathogen_targets = {}
pathogen_seq_pairs = {}

final = {}

cas_1 = 'SpCas9'
cas_2 = 'SpCas9'
orientation = 'PAM-out'

dist = system_selection_test(cas_1, cas_2, orientation, spacer = 'manual')

pathogen_pam_pairs['{}_{}_{}'.format(cas_1, cas_2, orientation)] = search_pam_pairs(pathogen_genomes, pathogen_pam, cas_1, cas_2, dist, orientation)
pathogen_targets['{}_{}_{}'.format(cas_1, cas_2, orientation)] = search_target_pairs(pathogen_pam_pairs, pathogen_genomes, cas_1, cas_2, orientation)
pathogen_seq_pairs['{}_{}_{}'.format(cas_1, cas_2, orientation)] = get_pathogen_seq_pairs(pathogen_pam_pairs, pathogen_targets, cas_1, cas_2, orientation)

final['{}_{}_{}'.format(cas_1, cas_2, orientation)] = final_for_one_pat(pathogen_dict, pathogen_targets, pathogen_pam_pairs, cas_1, cas_2, orientation)
```
